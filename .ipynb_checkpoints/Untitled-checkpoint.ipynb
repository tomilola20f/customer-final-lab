```python
from sklearn.preprocessing import MinMaxScaler

normalized = dfcopy.select_dtypes(np.number)
dfcopy[normalized.columns] = MinMaxScaler().fit_transform(normalized)
```

```python
dfcopy
```

```python
dfcopy= dfcopy.drop('Policy Type',axis=1)
dfcopy= dfcopy.drop('Customer',axis=1)
dfcopy= dfcopy.drop('Response',axis=1)
dfcopy= dfcopy.drop('Education',axis=1)
dfcopy= dfcopy.drop('Gender',axis=1)
dfcopy= dfcopy.drop('Policy',axis=1)
dfcopy= dfcopy.drop('Renew Offer Type',axis=1)
dfcopy= dfcopy.drop('Sales Channel',axis=1)
dfcopy= dfcopy.drop('Vehicle Class',axis=1)
dfcopy= dfcopy.drop('State',axis=1)
dfcopy= dfcopy.drop('Effective To Date',axis=1)

## the columns are not numericals that is why i drop them.
```

```python
dfcopy["Coverage"] = dfcopy["Coverage"].map({"Basic" : 0, "Extended" : 1, "Premium" : 2})
dfcopy["EmploymentStatus"] = dfcopy["EmploymentStatus"].map({"Employed" : 0, "Unemployed" : 1, "Medical Leave" : 2, "Disabled" : 3, "Retired" : 4})
dfcopy["Location Code"] = dfcopy["Location Code"].map({"Suburban" : 0, "Rural" : 1, "Urban" : 2})
dfcopy["Vehicle Size"] = dfcopy["Vehicle Size"].map({"Small" : 0, "Medsize" : 1, "Large" : 2})
#Encode the categorical variables
```

```python
dfcopy
```

```python
dfcopy= dfcopy.drop('Marital Status',axis=1)
```

```python
dfcopy = dfcopy.astype('float64')## changing it to numerical
```

```python
dfcopy
```

```python
dfcopy.corr() #checking for the correlation
```

```python
X=dfcopy.drop(['Total Claim Amount'], axis=1)
y=dfcopy['Total Claim Amount']
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.25, random_state = 42)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
```

```python
reg.fit(X_train, y_train)
```

```python
from sklearn.metrics import r2_score
predictions_train = reg.predict(X_train) # create predictions for our train data
predictions_test = reg.predict(X_test)   # create predictions for our test data

r2_train = r2_score(y_train, predictions_train) # calculate r2 score for train data
r2_test = r2_score(y_test, predictions_test)    # calculate r2 score for test data

print('R2 value for train: {}'.format(r2_train))
print('R2 value for test: {}'.format(r2_test)) 
```

```python
from sklearn.metrics import mean_squared_error

display(mean_squared_error(y_train,predictions_train))
display(mean_squared_error(y_test,predictions_test))

display(mean_squared_error(y_test,predictions_test,squared = False))
display(mean_squared_error(y_train,predictions_train,squared = False))
```

```python
## using the MLPRegressor
lin = LinearRegression()
knn = KNeighborsRegressor(n_neighbors = 5)
mlp = MLPRegressor(max_iter = 1000)
def model_inplace(model, X_train, X_test, y_train, y_test):
    model.fit(X_train,y_train)
    print("score on train set:", model.score(X_train, y_train))
    print("score on test set:", model.score(X_test, y_test))
model_inplace(lin, X_train, X_test, y_train, y_test)
model_inplace(knn, X_train, X_test, y_train, y_test)
model_inplace(mlp, X_train, X_test, y_train, y_test)
```
